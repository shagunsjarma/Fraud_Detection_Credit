```python
from sklearn.preprocessing import StandardScaler, RobustScaler

std_scaler = StandardScaler()
rob_scaler = RobustScaler()

df['normAmount'] = rob_scaler.fit_transform(df['Amount'].values.reshape(-1,1))
df['normTime'] = rob_scaler.fit_transform(df['Time'].values.reshape(-1,1))  
df.drop(['Time','Amount'], axis=1, inplace=True)
```

```python
scaled_amount = df['normAmount']
scaled_time = df['normTime']

df.drop(['normAmount','normTime'], axis=1, inplace=True)
df.insert(0, 'scaled_amount', scaled_amount)
df.insert(1, 'scaled_time', scaled_time)
df.head()
```

```python
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit, StratifiedKFold
```

```python
X = df.drop('Class', axis=1)
y = df['Class']

sss = StratifiedKFold(n_splits=5, shuffle=False, random_state=None)
for train_index, test_index in sss.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

original_Xtrain = X_train.values
original_Xtest = X_test.values
original_ytrain = y_train.values
original_ytest = y_test.values

```

```python
original_Xtrain.shape
```

```python
over_smapling = SMOTE(random_state=42)
```

```python
X_res, y_res = over_smapling.fit_resample(original_Xtrain, original_ytrain)
```

```python
print("Orginal Dataset shape:", Counter(original_ytrain))
```

```python
print("Dataset shape after oversampling:", Counter(y_res))
```

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "Support Vector Machine": SVC(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    }
```

```python
res = {}
for model_name, model in models.items():
    clf = model
    clf.fit(X_res, y_res)
    y_pred = clf.predict(original_Xtest)
    res[model_name] = classification_report_imbalanced(original_ytest, y_pred, output_dict=True)
    
```

```python
## best performing model
for model_name, metrics in res.items():
    print(f"Model: {model_name}")
    print(f"Precision (Fraud): {metrics['1']['pre']:.4f}")
    print(f"Recall (Fraud): {metrics['1']['rec']:.4f}")
    print(f"F1-Score (Fraud): {metrics['1']['f1']:.4f}")
    print("-" * 30)
```

```python
#predict pipeline for best model
def predict_fraud(input_data):
    #reshape the input data
    reshaped_data = np.array(input_data).reshape(1, -1)
    
    #preprocessing steps
    #scaled_data = rob_scaler.transform(reshaped_data)
    
    #prediction using the best model
    best_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
    best_model.fit(X_res, y_res)
    prediction = best_model.predict(reshaped_data)
    return prediction
    
    
```

```python
input_data = original_Xtest[0]
prediction = predict_fraud(input_data)
print("Prediction (0: No Fraud, 1: Fraud):", prediction)
print("Prediction (0: No Fraud, 1: Fraud):", original_ytest[0])
```
